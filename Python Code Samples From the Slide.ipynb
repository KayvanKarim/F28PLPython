````python
def tower(n):
    if n==0:
        return 2
    else:
        return 2**(tower(n-1))
````

### Reduce
`reduce(f, [a1,a2,a3,...,an])` computes `f(...f(f(a1,a2),a3),...,an)`

Reduce is a really useful function for performing some computation on a list and returning the result. It applies a rolling computation to sequential pairs of values in a list. For example, if you wanted to compute the product of a list of integers.<br>
`functools.reduce(func, iter, [initial_value])` cumulatively performs an operation on all the iterable’s elements and, therefore, can’t be applied to infinite iterables.

`func` must be a function that takes two elements and returns a single value. `functools.reduce()` takes the first two elements `A` and `B` returned by the iterator and calculates `func(A, B)`.<br>
It then requests the third element, `C`, calculates `func(func(A, B), C)`, combines this result with the fourth element returned, and continues until the iterable is exhausted. If the iterable returns no values at all, a TypeError exception is raised. If the initial value is supplied, it’s used as a starting point and `func(initial_value, A)` is the first calculation.
https://docs.python.org/3/howto/functional.html

Apply function of two arguments cumulatively to the items of iterable, from left to right, so as to reduce the iterable to a single value. For example, `reduce(lambda x, y: x+y, [1, 2, 3, 4, 5])` calculates `((((1+2)+3)+4)+5)`. The left argument, `x`, is the accumulated value and the right argument, `y`, is the update value from the iterable. If the optional initializer is present, it is placed before the items of the iterable in the calculation, and serves as a default when the iterable is empty. If initializer is not given and iterable contains only one item, the first item is returned.
https://docs.python.org/3/library/functools.html#functools.reduce

````python
product = 1
list = [1, 2, 3, 4]
for num in list:
    product = product * num
print(product)
````

````python
from functools import reduce
product = reduce((lambda x, y: x * y), [1, 2, 3, 4])
print(product)
````

# More on Recursion
[Back to TOC ↑](#table-of-content)

<font color="red">You are expected to know</font>
- Recursion is when a program calls itself. See the notes on Ocaml.
- Tail-recursion is when a program calls itself and this is the end of that program’s computation. Therefore, the local state of the program does not need to be stored, and indeed, it can be re-used. Tail-recursion is, in effect, equal to iteration.

https://python-course.eu/python3_recursive_functions.php

Let’s look at four implementations of Fibonacci:

````python
def fibi(n): # iterative
    a, b = 1, 1
    for i in range(n):
        a, b = b, a+b
    return(a)
````

````python
fibi(5)
````

````python
def fibr(n): # recursive 
    #print(n,end=' ')
    if n==0 or n==1:
        return(1)
    else:
        return fibr(n-1)+fibr(n-2)
````

````python
fibr(5)
````

````python
memo = { 0:1, 1:1 } # memoised
def fibm(n):
    if not n in memo:
        memo[n] = fibm(n-1)+fibm(n-2)
    return memo[n]
````

In computing, memoization or memoisation is an optimization technique used primarily to speed up computer programs by storing the results of expensive function calls and returning the cached result when the same inputs occur again.

https://en.wikipedia.org/wiki/Memoization

````python
fibm(5)
````

````python
def fibg(n): # with static local 
    if not hasattr(fibg, "memo"): #has attribute
        fibg.memo = { 0:1, 1:1 }
    if not n in fibg.memo:
        fibg.memo[n]=fibg(n-1)+fibg(n-2)
    return fibg.memo[n]
````

The arguments are an object and a string. The result is `True` if the string is the name of one of the object’s attributes, `False` if not. (This is implemented by calling `getattr(object, name)` and seeing whether it raises an `AttributeError` or not.)

https://docs.python.org/3/library/functions.html#hasattr

````python
fibg(5)
````

# Generators

[Back to TOC ↑](#table-of-content)

A method, containing a **`yield`** expression, is a **generator**.

Generator functions are like normal functions in most respects, and in fact are coded with normal def statements. However, when created, they are compiled specially into an object that supports the iteration protocol. 

The chief code difference between generator and normal functions is that a generator yields a value, rather than returning one—the `yield` statement suspends the function and sends a value back to the caller, but retains enough state to enable the function to resume from where it left off. When resumed, the function continues execution immediately after the last yield run. From the function’s perspective, this allows its code to produce a series of values over time, rather than computing them all at once and sending them back in something like a list.(From Learning Python, 5th Edition by Mark Lutz)

````python
def gensquares(n):
        for i in range(n):
            yield i ** 2
````

````python
f=gensquares(n)
print(f)
````

````python
for i in gensquares(5):     # Resume the function
        print(i , end=',')
````

````python
def reverse(data):
    for index in range(len(data)-1, -1, -1):
        yield data[index]
````

````python
for char in reverse('golf'): print (char , end='')
````
